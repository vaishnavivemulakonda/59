```python
import numpy as np
lst1=[1,2,3]
array1=np.array(lst1)
array1
```

```python
type(lst1)
```

```python
lst2=[4,5,6]
array2=np.array(lst2)
array2
```

```python
print("array2 multiplied by array1:",array1*array2)
print("array2 divided by array1:",array2/array1)
print("array2 raised to power of array1:",array2**array1)
print("array2 added by array1:",array1+array2)
print("array2 is subtracted by array1:",array2-array1)
```

```python
#sine function
print("sin:",np.sin(array1))
#logarithm
print("Natural logarithm:",np.log(array1))
print("Base-10 logarithm:",np.log10(array1))
print("Base-2 logarithm:",np.log2(array1))
#exponential
print("exponential:",np.exp(array1))
```

```python
people=['ann','arman','imran','nikki','remo','julie']
age=[21,23,14,16,18,17]
weight=[55,34,64,70,48,59]
height=[160,130,159,200,165,171]
```

```python
import matplotlib.pyplot as plt
plt.scatter(weight,height)
plt.title("Relationship between weight and height of patients")
plt.ylabel("height in meters")
plt.xlabel("weight of patients in kilograms")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.scatter(age,height)
plt.title("Relationship between age and height of patients")
plt.ylabel("height in meters")
plt.xlabel("age of patients")
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.title("people's weight in kgs",fontsize=18,fontstyle='italic')
#main plot function 'bar'
plt.bar(x=people,height=weight,width=0.6,color='orange',edgecolor='k',alpha=0.6)
plt.xlabel("people",fontsize=15)
plt.xtricks(fontsize=14,rotation=30)
plt.ytricks(fontsize=14)
plt.ylabel("weight(in kgs)",fontsize=15)
```
